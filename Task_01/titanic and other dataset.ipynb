```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Titanic dataset
url = '/content/tested.csv'
df = pd.read_csv(url)


df['Age'].fillna(df['Age'].median(), inplace=True)


df['Cabin'].fillna('Unknown', inplace=True)


df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

plt.figure(figsize=(8, 5))
gender_counts = df['Sex'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
plt.title('Gender Distribution of Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()


age_bins = [0, 12, 18, 30, 50, 100]
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)


age_group_counts = df['Age Group'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Age Distribution of Titanic Passengers')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd


csv_file = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
df = pd.read_csv(csv_file, skiprows=4)

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_2020 = df[['Country Name', '2020']].dropna()


df_2020_sorted = df_2020.sort_values(by='2020', ascending=False).head(20)

df_2020_sorted.plot(kind='bar', x='Country Name', y='2020', figsize=(12, 8), color='skyblue')


plt.xticks(rotation=45, ha='right')


plt.title('Top 20 Countries by Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population (in billions)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_2022 = df[['Country Name', '2022']].dropna()


df_2022_sorted = df_2022.sort_values(by='2022', ascending=False).head(20)


colors = plt.cm.tab20(np.linspace(0, 1, len(df_2022_sorted)))

df_2022_sorted.plot(kind='bar', x='Country Name', y='2022', figsize=(12, 8), color=colors)


plt.xticks(rotation=45, ha='right')


plt.title('Top 20 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population (in billions)')
plt.tight_layout()
plt.show()
```
